Description Of Task :We need to read the image in RBG format and then convert it to a grayscale image. This will turn an image into a classic black and white photo. **Invert the blurred image**

```python
invertblurimages = 255 - blurimages
plt.figure(figsize=(8,8))
plt.imshow(invertblurimages,cmap='gray')
plt.show()
```

```python
sketch = cv2.divide(grayimages, invertblurimages, scale=256.0)
plt.figure(figsize=(8,8))
plt.imshow(rgbimages)
plt.figure(figsize=(8,8))
plt.imshow(sketch,cmap='gray')
plt.show()
```
